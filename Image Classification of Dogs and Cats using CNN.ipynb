```python
import os
import tqdm
from keras.layers.convolutional import Convolution2D, MaxPooling2D
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Activation, Conv2D, MaxPooling2D
import cv2
import numpy as np
```

```python
train_directory = r'C:\Users\atjayade\Downloads\P14-Convolutional-Neural-Networks\Convolutional_Neural_Networks\dataset\training_set'
test_directory = r'C:\Users\atjayade\Downloads\P14-Convolutional-Neural-Networks\Convolutional_Neural_Networks\dataset\test_set'
```

```python
def label_image(img):
    label = img.split('.')[-3]
    if label == 'cat':
        return [1,0]
    if label == 'dog':
        return [0,1]
    
```

```python
# Creation of training data
# This part creates the training data by converting colour images to grayscale images.
# Gray scale images are easier to process

training=[]

for item1 in os.listdir(train_directory):
    path1 = os.path.join(train_directory, item1)
    for item in tqdm.tqdm(os.listdir(path1)):
        lab = label_image(item)
        img_path = os.path.join(path1, item)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (50,50))
        training.append([np.array(img), lab])
```

```python
from random import shuffle
```

```python
testing=[]

for item1 in os.listdir(test_directory):
    path1 = os.path.join(test_directory, item1)
    for item in tqdm.tqdm(os.listdir(path1)):
        lab = label_image(item)
        img_path = os.path.join(path1, item)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (50,50))
        testing.append([np.array(img), lab])
```

```python
shuffle(training)
training_ip=[]
training_op=[]
for item in training:
    training_ip.append(item[0])
    training_op.append(item[1])

shuffle(testing)
testing_ip=[]
testing_op=[]
for item in testing:
    testing_ip.append(item[0])
    testing_op.append(item[1])

training_ip = np.array(training_ip).reshape(-1, 50, 50, 1)
testing_ip = np.array(testing_ip).reshape(-1, 50, 50, 1)
```

```python
print(training_op)
```

```python
model = Sequential()
```

```python
model.add(Conv2D(filters=32, kernel_size=(3,3),padding='SAME', input_shape=training_ip[0].shape))

model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(filters=64, kernel_size=(3,3), padding='SAME'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(rate=0.4))

model.add(Conv2D(filters=128, kernel_size=(3,3), padding='SAME'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(rate=0.35))

model.add(Conv2D(filters=64, kernel_size=(3,3), padding='SAME'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(filters=64, kernel_size=(3,3), padding='SAME'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(1024))
model.add(Activation('relu'))
model.add(Dropout(rate=0.3))

model.add(Dense(2))
model.add(Activation('softmax'))
```

```python
import keras
opt = keras.optimizers.Adam()
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
print(model.summary())
```

```python
model.fit(training_ip, np.array(training_op),validation_data = (testing_ip, np.array(testing_op)), epochs=40, verbose=2)
```

```python
model.evaluate(testing_ip, np.array(testing_op), verbose=2)
```
